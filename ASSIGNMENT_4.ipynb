```
X_students = np.array(students.iloc[:,5].values)
y_test = np.array(students.iloc[:,6].values)
print( 'X_students' , X_students [:6] )
print( 'y_students' , y_students [:5] )
```

```
X_students = X_students.reshape( -1 ,1 )
print( 'X_students\n' , X_students [:5] )
```

```
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit( X_students, y_students )
model.score( X_students, y_students )
```

```
import matplotlib.pyplot as plt
plt.scatter( X_students , y_students )
print(X_students.shape)
plt.plot( X_students , model.predict( X_students.reshape( - 1 , 1 ) ) , 'r')
```
